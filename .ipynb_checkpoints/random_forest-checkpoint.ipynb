```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.utils import shuffle
from sklearn import metrics
from sklearn.metrics import plot_confusion_matrix
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerLine2D
import sys, os, time, csv, glob, cv2
from sklearn.svm import SVC

os.environ["KERAS_BACKEND"] = "plaidml.keras.backend"

pd.set_option('display.max_columns', None)
pd.options.display.float_format = '{:.5f}'.format
np.set_printoptions(threshold=sys.maxsize)
np.set_printoptions(precision = 5, suppress = True)
```

```python
attack_test = pd.read_csv('/Users/ryan/NSL CNN/2d-cnn/datasets/80-20/attack_test_8020.csv')
attack_train = pd.read_csv('/Users/ryan/NSL CNN/2d-cnn/datasets/80-20/attack_train_8020.csv')
normal_test = pd.read_csv('/Users/ryan/NSL CNN/2d-cnn/datasets/80-20/normal_test_8020.csv')
normal_train = pd.read_csv('/Users/ryan/NSL CNN/2d-cnn/datasets/80-20/normal_train_8020.csv')
```

```python
train_df = pd.merge(attack_train, normal_train, how='outer')
```

```python
test_df = pd.merge(attack_test, normal_test, how='outer')
```

```python
train_df['attack_type'] = np.where(train_df.attack_type == 'attack', 0, train_df.attack_type)
train_df['attack_type'] = np.where(train_df.attack_type == 'normal', 1, train_df.attack_type)
test_df['attack_type'] = np.where(test_df.attack_type == 'attack', 0, test_df.attack_type)
test_df['attack_type'] = np.where(test_df.attack_type == 'normal', 1, test_df.attack_type)
```

```python
train_df
```

```python
train_df.attack_type = train_df.attack_type.astype(int)
test_df.attack_type = test_df.attack_type.astype(int)
```

```python
train_df = shuffle(train_df, random_state = 7)
test_df = shuffle(test_df, random_state = 7)
```

```python
y_train = train_df['attack_type']
y_test = test_df['attack_type']

train_df = train_df.drop('attack_type', axis = 1)
test_df = test_df.drop('attack_type', axis = 1)
```

```python
rf = RandomForestClassifier(n_estimators = 50, random_state = 42)
rf.fit(train_df, y_train)
```

```python
preds = rf.predict(test_df)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, roc_curve, auc
CLASS = ['Attack', 'Normal']
```

```python
check_element = 5
result = np.where(preds == check_element)

print(f'First n predictions:\n\n {(preds[:10])}\n')
print(f'Category classification of element at index {check_element}: {CLASS[np.argmax(preds[check_element])]}')
```

```python
print(f'Accuracy Score: {round(accuracy_score(y_test, preds), 5)*100}%\n\nClassification Table:\n{classification_report(y_test, preds, target_names = CLASS)}')
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
plt.rcParams.update({'font.size': 16})
plot_confusion_matrix(rf, test_df, y_test, ax = ax, display_labels = CLASS)
```

```python
false_pos_rate, true_pos_rate, thresholds = roc_curve(y_test, preds)
roc_auc = auc(false_pos_rate, true_pos_rate)*100

round(roc_auc, 5)
```

```python
n_estimators = [1,2,5,10,25,50,100,150,200]
train_results = []
test_results = []

for estimator in n_estimators:
    rf = RandomForestClassifier(n_estimators = estimator, n_jobs = -1)
    rf.fit(train_df, y_train)
    train_pred = rf.predict(train_df)
    false_pos_rate, true_pos_rate, thresholds = roc_curve(y_train, train_pred)
    roc_auc = auc(false_pos_rate, true_pos_rate)
    train_results.append(roc_auc)
    y_pred = rf.predict(test_df)
    false_pos_rate, true_pos_rate, thresholds = roc_curve(y_test, y_pred)
    roc_auc = auc(false_pos_rate, true_pos_rate)
    test_results.append(roc_auc)

```

```python
plt.figure(figsize=(24,12))
line1, = plt.plot(n_estimators, train_results, 'orange', label = 'Train AUC', lw = 4)
line2, = plt.plot(n_estimators, test_results, 'purple', label = 'Test AUC', lw = 4)
plt.legend(handler_map = {line1: HandlerLine2D(numpoints = 5)}, loc = 'best', fontsize = 16)
plt.title('NSL-KDD Random Forest', fontsize = 24)
plt.ylabel('AUC Score', fontsize = 24)
plt.xlabel('n_estimators', fontsize = 24)
plt.xticks(np.arange(0, 210, step = 10))
plt.grid(axis = 'x')
plt.show()
```

```python
display = metrics.RocCurveDisplay(fpr = false_pos_rate, tpr = true_pos_rate, roc_auc = roc_auc, estimator_name = 'NSL-KDD RF')
display.plot()  
plt.show()
```
