```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import sys, os, time

# Settings
pd.set_option('display.max_columns', None)
pd.options.display.float_format = '{:.5f}'.format
np.set_printoptions(threshold=sys.maxsize)
np.set_printoptions(precision = 5, suppress = True)
```

```python
df = pd.read_csv('datasets/combined_dataset.csv')
df
```

```python
normal_df = df.loc[df.attack_type == 'normal']
attack_df = df.loc[df.attack_type == 'attack']
```

```python
attack_df
```

```python
def dataframe_splitter(dataframe, split = float):
    df = dataframe
    
    train_df = df.sample(frac = split, random_state = 7)
    test_df = df.drop(train_df.index)
    
    return train_df, test_df

split = 0.8
attack_train, attack_test = dataframe_splitter(attack_df, split = split)
```

```python
normal_train
```

```python
attack_train.to_csv(f'datasets/80-20/attack_train_{int(split*100)}{int(100-split*100)}.csv', index = 0)
attack_test.to_csv(f'datasets/80-20/attack_test_{int(split*100)}{int(100-split*100)}.csv', index=0)
```
